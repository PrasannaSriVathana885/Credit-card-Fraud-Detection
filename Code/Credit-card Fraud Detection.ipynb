2. Check for data imbalance

```python
df['Class'].value_counts()
```

```python
import matplotlib.pyplot as plt

class_counts = df['Class'].value_counts()
class_counts.plot(kind='bar')
plt.title('Original Class Distribution')
plt.xlabel('Class (0 = Not Fraud, 1 = Fraud)')
plt.ylabel('Count')
plt.show()
```

2.1 Undersampling

```python
from sklearn.utils import resample
df_majority = df[df.Class==0]
df_minority = df[df.Class==1]
df_majority_downsampled = resample(df_majority, 
                                   replace=False,     
                                   n_samples=len(df_minority), 
                                   random_state=42)
df_balanced = pd.concat([df_majority_downsampled, df_minority])
```

```python
df_balanced['Class'].value_counts()
```

```python
class_counts = df_balanced['Class'].value_counts()
class_counts.plot(kind='bar')
plt.title('Original Class Distribution')
plt.xlabel('Class (0 = Not Fraud, 1 = Fraud)')
plt.ylabel('Count')
plt.show()
```

2.2 Oversampling

```python
df_minority_upsampled = resample(df_minority, 
                                 replace=True,     
                                 n_samples=len(df_majority), 
                                 random_state=42)
df_balanced = pd.concat([df_majority, df_minority_upsampled])
```

```python
df_balanced['Class'].value_counts()
```

```python
class_counts = df_balanced['Class'].value_counts()
class_counts.plot(kind='bar')
plt.title('Original Class Distribution')
plt.xlabel('Class (0 = Not Fraud, 1 = Fraud)')
plt.ylabel('Count')
plt.show()
```

```python
X = df.drop(columns=['Class'])
y = df['Class']
```

2.3 Synthetic Data Sampling using SMOTE 

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)
```

3. Feature Engineering

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X_res)
```

```python
import seaborn as sns
plt.figure(figsize=(14, 10))
corr = df.corr()
sns.heatmap(corr, cmap='coolwarm', annot=False)
plt.title('Correlation Heatmap of Features')
plt.show()
```

4. Model Training and Selection

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_res, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, auc, roc_curve

lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
lr_pred = lr_model.predict(X_test)
print(classification_report(y_test, lr_pred))
disp = ConfusionMatrixDisplay(confusion_matrix(y_test, lr_pred))
disp.plot(cmap='Blues')
```

```python
from sklearn.svm import SVC
svc_model = SVC()
svc_model.fit(X_train, y_train)
svc_pred = svc_model.predict(X_test)
print(classification_report(y_test, svc_pred))
disp = ConfusionMatrixDisplay(confusion_matrix(y_test, svc_pred))
disp.plot(cmap='Blues')
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
print(classification_report(y_test, rf_pred))
disp = ConfusionMatrixDisplay(confusion_matrix(y_test, rf_pred))
disp.plot(cmap='Blues')
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Logistic Regression: use predict_proba
fpr, tpr, _ = roc_curve(y_test, lr_model.predict_proba(X_test)[:, 1])
roc_auc = auc(fpr, tpr)

# SVM: use decision_function
fpr1, tpr1, _ = roc_curve(y_test, svc_model.decision_function(X_test))
roc_auc1 = auc(fpr1, tpr1)

# Random Forest: use predict_proba
fpr2, tpr2, _ = roc_curve(y_test, rf_model.predict_proba(X_test)[:, 1])
roc_auc2 = auc(fpr2, tpr2)

# Plotting
plt.figure()
plt.plot(fpr, tpr, label=f'Logistic Regression (AUC = {roc_auc:.2f})')
plt.plot(fpr1, tpr1, label=f'SVM (AUC = {roc_auc1:.2f})')
plt.plot(fpr2, tpr2, label=f'Random Forest (AUC = {roc_auc2:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.title('Receiver Operating Characteristic')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.show()
```
